**1.** Given a random array, change the sign of the elements whose values ​​are between 3 and 8

```python
#Mathematically, we are trying to write the values within the range 3<= x<= 8.
#That is if the element x is between 3 and 8 inclusive, multiply by -1. Otherwise keep the value unchanged.

arr= np.array([1, 4, 7, 10, 3, 8, 6, 2])
arr[(arr > 3) & (arr < 8)] *= -1
print(arr)
```

**4.** Given two arrays A (8x3) and B (2x2), find the rows in A that contain elements from each row in B, regardless of the order of the elements in B.

```python
# Example arrays
A = np.array([
    [1, 2, 3],
    [4, 5, 6],
    [1, 4, 2],
    [7, 8, 9],
    [2, 1, 5],
    [6, 4, 5],
    [3, 1, 2],
    [8, 7, 6]
])

B = np.array([
    [1, 2],
    [4, 5]
])

# Find rows in A containing elements of each row in B
result = []

for b in B:
    mask = np.all(np.isin(b, A), axis=0)
    rows = [i for i, row in enumerate(A) if np.all(np.isin(b, row))]
    result.append(rows)

print(result)
```

**5.** Given a 10x3 matrix, find rows of unequal values ​​(for example, row [2,2,3] remains, row [3,3,3] is deleted)

```python
# Example 10x3 matrix
A = np.array([
    [2,2,3],
    [3,3,3],
    [4,5,6],
    [7,7,7],
    [1,2,1],
    [9,9,8],
    [5,5,5],
    [3,4,3],
    [6,6,7],
    [8,8,8]
])

# Keep rows where elements are not all equal
result = A[~np.all(A == A[:, [0]], axis=1)]

print(result)
```

**6.** Given a two-dimensional array, remove those lines that are repeated

* __Problem 1__: Calculate the product of nonzero elements on the diagonal of a rectangular matrix. For example, for X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) the answer is 3.

```python
X = [[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]
# Without Numpy
import math
result = math.prod(X[i][i] for i in range(min(len(X), len(X[0]))) if X[i][i] != 0)
#print(result)

# With Numpy
result = np.prod(np.diag(X)[np.diag(X) != 0])
print(result)
```

* __Problem 2__: Given two vectors x and y, check whether they define the same multiset. For example, for x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]), the answer is True.

```python
# Without Numpy

x = [1,2,2,4]
y = [4,2,1,2]

from collections import Counter
result = Counter(x) == Counter(y)
#print(result)

#With Numpy

result = np.array_equal(np.sort(x), np.sort(y))
print(result)
```

* __Problem 3__: Find the maximum element in the vector x among elements preceded by zero. For example, for x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) the answer is 5.

```python
x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])
# Without Numpy
result = max(x[i] for i in range(1, len(x)) if x[i-1] == 0 and x[i] != 0)
#print(result)

# With Numpy
result = np.max(x[1:][x[:-1] == 0])
print(result)
```

* __Problem 4__: Implement run-length encoding. For a given vector x, return a tuple of two vectors of the same length. The first contains numbers, and the second contains the number of times they should be repeated. For example, for x = np.array([2, 2, 2, 3, 3, 3, 5]), the answer is (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
x = np.array([2, 2, 2, 3, 3, 3, 5])

# Without Numpy

values, counts = [], []
for v in x:
    if not values or v != values[-1]:
        values.append(v); counts.append(1)
    else:
        counts[-1] += 1
# print(values, counts)

#With Numpy

idx = np.flatnonzero(np.r_[True, x[1:] != x[:-1]])
values = x[idx]
counts = np.diff(np.r_[idx, len(x)])
print(values, counts)
```

5.
